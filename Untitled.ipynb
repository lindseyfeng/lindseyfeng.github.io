```python
import matplotlib.pyplot as plt
import pandas as pd

# Data preparation
data = {
    "Model": [
        "pih3-base-1epoch", "pih3-base-2epoch", "phi3-base-3epoch", 
        "phi3-base-4epoch", "phi3-iterative-1epoch", "phi3-iterative-2epoch", 
        "phi3-iterative-3epoch", "phi3-iterative-4epoch"
    ],
    "Win Rate": [
        0.6567, 0.7435897436, 0.7160493827, 
        0.7126436782, 0.65625, 0.7532467532, 
        0.8048780488, 0.7692307692
    ],
    "Count": [94, 78, 81, 87, 64, 77, 82, 78]
}

df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 5))
for label, group_df in df.groupby(df['Model'].str.contains('iterative')):
    plt.plot(group_df['Model'], group_df['Win Rate'], marker='o', label='Iterative' if label else 'Base')

plt.title('Trend of Win Rates by Model Type')
plt.xlabel('Model')
plt.ylabel('Win Rate')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
